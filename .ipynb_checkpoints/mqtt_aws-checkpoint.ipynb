```python
from __future__ import print_function
import sys
import ssl
import logging, traceback
import paho.mqtt.client as mqtt
import ipywidgets as widgets

IoT_protocol_name = "x-amzn-mqtt-ca"
aws_iot_endpoint = "a1fuz731sf8v8p-ats.iot.us-east-1.amazonaws.com" # <random>.iot.<region>.amazonaws.com

ca = "credenciales/certificado-ca.pem" 
cert = "credenciales/certificado.pem.crt"
private = "credenciales/llave-privada.pem.key"

clients_connected = []
allTemps=[]

logger = logging.getLogger()
logger.setLevel(logging.DEBUG)
handler = logging.StreamHandler(sys.stdout)
log_format = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(log_format)
logger.addHandler(handler)

def ssl_alpn():
    try:
        # debug print opnessl version
        # logger.info("open ssl version:{}".format(ssl.OPENSSL_VERSION))
        ssl_context = ssl.create_default_context()
        ssl_context.set_alpn_protocols([IoT_protocol_name])
        ssl_context.load_verify_locations(cafile=ca)
        ssl_context.load_cert_chain(certfile=cert, keyfile=private)

        return  ssl_context
    except Exception as e:
        print("error con las credenciales")
        raise e

def on_connect(client, userdata, flags, rc):
    global clients_connected
    clients_connected.append(client)
    print(f'cliente connectado {client._client_id}')
    mqttc = mqtt.Client()
        

def listar_clientes():
    return list(map(lambda c: c._client_id.decode('utf-8'), clients_connected))
    

def on_message(client, userdata, msg):
    message=msg.payload.decode("utf-8")
    print('mensaje recibido:',message)

    
def on_publish(client, userdata, mid):
    print('publicado:',mid)


def crear_cliente(id, endpoint):
    cliente = mqtt.Client(id)
    ssl_context= ssl_alpn()
    cliente.tls_set_context(context=ssl_context)

    cliente.on_connect = on_connect
    cliente.on_message = on_message
    
    cliente.connect(endpoint, port=443)
    cliente.loop_start()
    
    return cliente

def subscribir_cliente(cliente, topico):
    cliente.subscribe(topico)

def publicar(cliente, topico, mensaje):
    cliente.publish(topico, mensaje)

def widget_publicar_demo(cliente):
    # temperatura a seleccionar con el slider
    temp=0
    
    # setea la temperatura desde el slider
    def setTemp(change):
        global temp
        temp=change['new']

    def publish(*args):
        global temp
        message = f'{{"temperatura":{temp}, "mensaje":"{text_box.value}"}}'
        cliente.publish("topico/prueba",message)
        
    button = widgets.Button(description='publicar temperatura!')
    slide = widgets.IntSlider(min=1, max=100, step=1, description='temperatura')
    text_box = widgets.Text(description='mensaje')
    ui = widgets.HBox([slide, text_box])

    slide.observe(setTemp, names='value')

    display(ui)
    display(button)
    button.on_click(publish)
```
